## ORDER BY

Panda sorts rows in ascending order by default. To reverse that, provide `ascending = False`

```python
# select * from airport_freq where airport_ident = 'KLAX' order by type
airport_frequencies[airport_frequencies.airport_ident == "KLAX"].sort_values("type")
```

```python
# select * from airport_freq where airport_ident = 'KLAX' order by type
airport_frequencies[airport_frequencies.airport_ident == "KLAX"].sort_values(
    "type", ascending=False
)
```

# GROUP BY, COUNT, ORDER BY

In pandas, `.count()` will return the number of non-null/NaN values. To get the same result as SQL `count` which ignores nulls, use `.size()`

```python
# select iso_country, type, count(*) from airports group by iso_country, type order by iso_country, type
airports.groupby(["iso_country", "type"]).size()
```

```python
# use .reset_index() to flatten the result.
airports.groupby(["iso_country", "type"]).size().reset_index()
```

```python
# select iso_country, type, count(*)
# from airports group
# group by iso_country, type
# order by iso_country, count(*) desc

# .to_frame('size') basically means convert this series to dataframe with the name size
# https://stackoverflow.com/questions/56920021/why-use-to-frame-before-reset-index
airports.groupby(["iso_country", "type"]).size().to_frame(
    "size"
).reset_index().sort_values(["iso_country", "size"], ascending=[True, False])
```

# HAVING

```python
# select type, count(*)
# from airports
# where iso_country = 'US'
# group by type
# having count(*) > 1000
# order by count(*) desc

airports[airports.iso_country == "US"].groupby("type").filter(
    lambda g: len(g) > 1000
).groupby("type").size().sort_values(ascending=False)
```

## Top N records

```python
by_country = (
    airports.groupby("iso_country").size().to_frame("airport_count").reset_index()
)


# select iso_country from by_country order by size desc limit 10
by_country.sort_values("airport_count", ascending=False).head(10)
```

```python
by_country.nlargest(10, columns="airport_count")
```

```python
# select iso_country from by_country order by size desc limit 10 offset 10
by_country.nlargest(20, columns="airport_count").tail(10)
```

# JOIN

Use `.merge()` to join pandas dataframes.

- `inner` (default)
- `left` is LEFT OUTER JOIN in SQL
- `right` is RIGHT OUTER in SQL
- `outer` is FULL OUTER in SQL

```python
# select airport_ident, type, description, frequency_mhz
# from airport_frequencies
# join airports on (airport_frequencies.airport_ref = airports.id)
# where airports.ident = 'KLAX'

airport_frequencies.merge(
    airports[airports.ident == "KLAX"][["id"]],
    left_on="airport_ref",
    right_on="id",
    how="inner",
)[["airport_ident", "type", "description", "frequency_mhz"]]
```

# UNION ALL and UNION

Use `pd.concat()` to `union all` two dataframes.

```python
# select name, municipality
# from airports
# where ident = 'KLAX'
#   union all
# select name, municipality
# from airports
# where ident = 'KLGB'

pd.concat(
    [
        airports[airports.ident == "KLAX"][["name", "municipality"]],
        airports[airports.ident == "KLGB"][["name", "municipality"]],
    ]
)
```

To drop duplicates, use `.drop_duplicates()`

```python
# select name, municipality
# from airports
# where ident = 'KLAX'
#   union
# select name, municipality
# from airports
# where ident = 'KLGB'

pd.concat(
    [
        airports[airports.ident == "KLAX"][["name", "municipality"]],
        airports[airports.ident == "KLAX"][
            ["name", "municipality"]
        ],  # Intentionally duplicate
        airports[airports.ident == "KLGB"][["name", "municipality"]],
    ]
).drop_duplicates()
```
